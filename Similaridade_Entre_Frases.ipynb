# Carregar conjunto de dados

```python
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
dataset = load_dataset("squad_v1_pt") # este é um dataset de perguntas e respostas
```

```python
# Vamos avaliar quais perguntas são similares
questions = dataset['train']['question']
```

```python
# vamos olhar so as 5 primeiras perguntas
for i in range(5):
  print(questions[i])
```

# Olhar a similaridade utilizando um modelo multilíngua

```python
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')
```

```python
# vamos codificar apenas 100 primeiras frases do nosso dataset apenas para
# exemplificar. Teste você com todo dataset e veja o tempo de execução.

# codificando as frases para obter suas representações
questions100_embeddings = model.encode(questions[:100])
```

```python
# Calculando a similaridade cosseno entre a  primeira frase e o restante das frases
similarity_scores = cosine_similarity([questions100_embeddings[0]], questions100_embeddings[1:])
```

```python
idx_max = np.argmax(similarity_scores) # aqui pegamos a posicao do elemento com o maior valor de similaridade
```

```python
similarity_scores[0][idx_max] # aqui vemos o valor da similaridade entre a frase em idx_max
```

```python
# agora vamos ver a frase em idx_max e na posição 0
print(questions[0])
print(questions[idx_max])
```

# Olhar a similaridade utilizando um modelo exclusivamente treinado para Português

```python
model_pt = SentenceTransformer('ricardo-filho/bert-base-portuguese-cased-nli-assin-2')
```

```python
# codificando as frases para obter suas representações
questions100_embeddings_pt = model_pt.encode(questions[:100])
```

```python
# Calculando a similaridade cosseno entre a  primeira frase e o restante das frases
similarity_scores_pt = cosine_similarity([questions100_embeddings_pt[0]], questions100_embeddings_pt[1:])
```

```python
idx_max_pt = np.argmax(similarity_scores_pt)
```

```python
similarity_scores_pt[0][idx_max_pt]
```

```python
# agora vamos ver a frase em idx_max e na posição 0
print(questions[0])
print(questions[idx_max_pt])
```
